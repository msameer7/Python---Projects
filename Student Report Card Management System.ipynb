```python
# Student Report Card Management System

students = []  # List to hold all student records

def add_student():
    student = {}
    student['id'] = input("Enter student ID: ")
    student['name'] = input("Enter student name: ")
    student['marks'] = []

    for subject in ['Math', 'English', 'Science']:
        mark = float(input(f"Enter marks for {subject}: "))
        student['marks'].append(mark)

    student['average'] = sum(student['marks']) / len(student['marks'])

    # Grade assignment logic
    if student['average'] >= 90:
        student['grade'] = 'A+'
    elif student['average'] >= 80:
        student['grade'] = 'A'
    elif student['average'] >= 70:
        student['grade'] = 'B'
    elif student['average'] >= 60:
        student['grade'] = 'C'
    else:
        student['grade'] = 'F'

    students.append(student)
    print("‚úÖ Student added successfully!\n")


def view_all_students():
    if not students:
        print("‚ö†Ô∏è No student records found.\n")
        return

    print("\nüìÑ All Student Records:")
    for s in students:
        print(f"ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}, Avg: {s['average']:.2f}, Grade: {s['grade']}")
    print()


def search_student():
    key = input("Enter student ID or name to search: ").lower()
    found = False
    for s in students:
        if s['id'].lower() == key or s['name'].lower() == key:
            print(f"üîç Found: ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}, Avg: {s['average']:.2f}, Grade: {s['grade']}\n")
            found = True
            break
    if not found:
        print("‚ùå Student not found.\n")


def main_menu():
    while True:
        print("===== Student Report Card System =====")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            view_all_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            print("üëã Exiting the system. Goodbye!")
            break
        else:
            print("‚ùó Invalid choice. Please try again.\n")


# Run the system
main_menu()
# Student Report Card Management System

students = []  # List to hold all student records

def add_student():
    student = {}
    student['id'] = input("Enter student ID: ")
    student['name'] = input("Enter student name: ")
    student['marks'] = []

    for subject in ['Math', 'English', 'Science']:
        mark = float(input(f"Enter marks for {subject}: "))
        student['marks'].append(mark)

    student['average'] = sum(student['marks']) / len(student['marks'])

    # Grade assignment logic
    if student['average'] >= 90:
        student['grade'] = 'A+'
    elif student['average'] >= 80:
        student['grade'] = 'A'
    elif student['average'] >= 70:
        student['grade'] = 'B'
    elif student['average'] >= 60:
        student['grade'] = 'C'
    else:
        student['grade'] = 'F'

    students.append(student)
    print("‚úÖ Student added successfully!\n")


def view_all_students():
    if not students:
        print("‚ö†Ô∏è No student records found.\n")
        return

    print("\nüìÑ All Student Records:")
    for s in students:
        print(f"ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}, Avg: {s['average']:.2f}, Grade: {s['grade']}")
    print()


def search_student():
    key = input("Enter student ID or name to search: ").lower()
    found = False
    for s in students:
        if s['id'].lower() == key or s['name'].lower() == key:
            print(f"üîç Found: ID: {s['id']}, Name: {s['name']}, Marks: {s['marks']}, Avg: {s['average']:.2f}, Grade: {s['grade']}\n")
            found = True
            break
    if not found:
        print("‚ùå Student not found.\n")


def main_menu():
    while True:
        print("===== Student Report Card System =====")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_student()
        elif choice == '2':
            view_all_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            print("üëã Exiting the system. Goodbye!")
            break
        else:
            print("‚ùó Invalid choice. Please try again.\n")


# Run the system
main_menu()
```
